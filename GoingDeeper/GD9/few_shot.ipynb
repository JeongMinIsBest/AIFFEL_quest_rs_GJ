```python
import json

# 기존 파일 불러오기
with open("fewshot_examples.json", "r", encoding="utf-8") as f:
    data = json.load(f)

converted = []

for category, examples in data.items():
    for ex in examples:
        prompt = ex["user"]
        completion = ex["model"]
        tokens = len(completion)  # 글자 수를 tokens로 사용
        converted.append({
            "prompt": prompt,
            "completion": completion,
            "tokens": tokens
        })

# 새 파일로 저장
with open("fewshot_examples_converted.json", "w", encoding="utf-8") as f:
    json.dump(converted, f, ensure_ascii=False, indent=2)
```
